```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import math
import matplotlib.pyplot as plt
```

```python
# simulation data_3
u1 = np.hstack((5,np.zeros(19)))
u2 = np.hstack((0,5,np.zeros(18)))
u3 = np.hstack((7,np.zeros(19)))
cov_matrix=np.eye(20)
cluster_1 = np.random.multivariate_normal(u1,cov_matrix,1500)
cluster_2 = np.random.multivariate_normal(u2,cov_matrix,2000)
cluster_3 = np.random.multivariate_normal(u3,cov_matrix,1500)
data = np.vstack((cluster_1,cluster_2,cluster_3))
label = np.hstack((np.ones(1500),np.ones(2000)*2,np.ones(1500)*3)).reshape([5000,1])
```

```python
tsne = TSNE()
reslut = pd.DataFrame(tsne.fit_transform(data))
d = reslut[label==1]
plt.plot(d[0],d[1],'ro')
d = reslut[label==2]
plt.plot(d[0],d[1],'go')
d = reslut[label==3]
plt.plot(d[0],d[1],'bo')
plt.show()
```

```python
#写入excel
data_output = pd.DataFrame(np.hstack((data,label)))
data_output.to_excel('simu_data_5000*21.xlsx')
```
